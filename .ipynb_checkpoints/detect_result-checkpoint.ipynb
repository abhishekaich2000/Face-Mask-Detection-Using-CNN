```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import cv2
from tensorflow import keras
```

```python
model = tf.keras.models.load_model('my_model.h5')
```

# Checking the predictions

```python
frame=cv2.imread('test2.png')
```

```python
plt.imshow(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
```

```python
final_image=cv2.resize(frame, (224,224))
final_image=np.expand_dims(final_image, axis=0)
final_image=final_image/255.0
```

```python
predictions=model.predict(final_image)
predictions
```

```python
if predictions>0.1:
    print("No mask detected")
else:
    print("Mask detected")
```

```python
face_classifier = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")

video_capture = cv2.VideoCapture("without-mask.mp4")

while True:
    # Capture frame-by-frame
    ret, img = video_capture.read()

    img_gray=cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
    face=face_classifier.detectMultiScale(
        img_gray,
        scaleFactor=1.1,
        minNeighbors=5,
        minSize=(30, 30),
    )
    for (x,y,w,h) in face:
        final_image=img[y:y+w,x:x+w]
        final_image=np.expand_dims(final_image, axis=0)
        final_image=final_image/255.0
        prediction=model.predict(final_image);
        print(prediction)
        if(prediction > 0.9):
            cv2.rectangle(img, (x,y), (x+w,y+h), (0,0,255), 2)
            cv2.rectangle(img,(x,y-40),(x+w,y), (0,0,255),-1)
            cv2.putText(img, "No Mask", (x, y-60),cv2.FONT_HERSHEY_SIMPLEX,4,(255,255,255),6)
        else:
            cv2.rectangle(img, (x,y), (x+w,y+h), (0, 255, 0), 2)
            cv2.rectangle(img,(x,y-40),(x+w,y), (0,255,0),-1)
            cv2.putText(img, "Mask", (x, y-50),cv2.FONT_HERSHEY_SIMPLEX,4,(255,255,255),4)
            
    cv2.imshow('video',img)          

    key=cv2.waitKey(1)
    
    if(key==27):
        break  
        

# When everything is done, release the capture
video_capture.release()
cv2.destroyAllWindows()

```
